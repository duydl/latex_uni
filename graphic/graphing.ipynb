```python
import matplotlib.pylab as plt
import matplotlib
import pandas as pd
import numpy as np
```

```python
data1 = np.array(pd.read_csv("kobe_ns.dat", sep='\s+', header = None)).T
data1
```

```python
fig, ax = plt.subplots()
ax.plot(data1[0], data1[1])
ax.set(xlabel='Elapsed time (s)', ylabel='NS Acceleration (cm/$s^2$)', title='Great Hanshin Earthquake (NS direction), JMA Record')
plt.show()
plt.savefig('1.svg', format='svg')
```

```python
import PyGnuplot as gp
gp.default_term = 'wxt'
X = np.arange(10)
Y = np.sin(X/(2*np.pi))
Z = Y**2.0
gp.s([X,Y,Z])
gp.c('splot "calxyz.dat"')
gp.p('myfigure.ps')
```
